# Inspecting 3D image data
This notebook demonstrates how to navigate through 3D images.

```python
import numpy as np
from skimage.io import imread, imshow
import stackview
```

```python
# Laod example data
mri_image = imread('data/Haase_MRT_tfl3d1.tif')
mri_image.shape
```

## Slicing
As also 3D images are just arrays, we can use the same syntax to select planes as if we were selecting array elements.

```python
xy_plane = mri_image[100]

imshow(xy_plane)
```

```python
zx_plane = mri_image[:,128,:]

imshow(zx_plane)
```

You can transpose axes using the `.T` transpose operation.

```python
imshow(zx_plane.T)
```

## Interactive slicing
The [stackview](https://github.com/haesleinhuepf/stackview) library allows interactive slicing.

```python
stackview.slice(mri_image)
```

## Projection
Projections along an axis allow to visualize 3D stacks as well. When specifying axes in Python, typically Z is the first axis (`0`), Y is the second (`1`) and X the last (`2` or `-1`)

```python
# Maximum Intensity Projection along Z
projection = mri_image.max(axis=0)

# show result
stackview.imshow(projection)
```

The average or "mean" projection appears more _transparent_:

```python
# Mean Z Projection
projection = mri_image.mean(axis=0)

# show result
imshow(projection)
```

```python
# Maximum X Projection
projection = mri_image.max(axis=1)

# show result
imshow(projection.T)
```
