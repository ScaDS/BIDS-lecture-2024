```python
mesh = nppas.to_vedo_mesh(surface)

filename = "data/branchoid.ply"

_ = vedo.write(mesh, filename)
```

## Loading meshes from disk

```python
new_mesh = vedo.load(filename)
new_mesh
```

```python
new_surface = nppas.to_napari_surface_data(new_mesh)
new_surface
```
