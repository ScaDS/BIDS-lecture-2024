# Cupy as drop-in replacement for numpy
[Cupy](https://docs.cupy.dev/) can be used as a drop-in replacement for [numpy]()

```python
import stackview
```

```python
try:
    import cupy as xp
except:
    import numpy as xp

import numpy as np

xp
```

```python
from skimage.io import imread, imshow
```

```python
image = imread("../03b_image_processing/data/blobs.tif")
```

```python
imshow(image)
```

```python
xp_image = xp.asarray(image)
```

Cupy images cannot be shown using scikit-image's `imshow`. This line would lead to an error:

```python
# imshow(xp_image)
```

We need to convert the image back to a numpy array before showing it.

```python
if np == xp:
    np_image = xp_image
else:
    np_image = xp.asnumpy(xp_image)

imshow(np_image)
```

Stackview's imshow takes care of that under the hood.

```python
stackview.insight(xp_image)
```
